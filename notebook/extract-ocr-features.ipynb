```python
# Tạo folder
%cd /content
!mkdir Images_new
!mkdir unzip_images
!mkdir my_images
!mkdir ocr_features

# Unzip các file vào folder vừa tạo (Sửa cái path ở đây)
!unzip '/content/drive/MyDrive/UIT/DS201_DO_AN/ViTextCap Data/Raw Data/Images_new.zip' -d Images_new


# Extract toàn bộ file zip chứa ảnh
import os
import glob
import shutil
import zipfile

zipfiles_ = glob.glob('Images_new/Data/*.zip')

for z_file in zipfiles_:
  with zipfile.ZipFile(z_file, 'r') as zip_ref:
    zip_ref.extractall('unzip_images')

# Chia ảnh
all_images = glob.glob('/content/unzip_images/*')
n_images = len(all_images)
n_images_1_4 = int(n_images * 0.25)
n_images_1_2 = int(n_images * 0.5)
n_images_3_4 = int(n_images * 0.75)

member_images = {'Đuc_Anh': all_images[0: n_images_1_4],
                 'Hien': all_images[n_images_1_4: n_images_1_2],
                 'Huy': all_images[n_images_1_2: n_images_3_4],
                 'Nhi': all_images[n_images_3_4: n_images]}

###### SỬA TÊN Ở ĐÂY ###### (Sửa xong cái tên thì còn lại giữ nguyên hết)
member_name = 'Hien'
for i in member_images[member_name]:
  shutil.copy(i, '/content/my_images')

# len(os.listdir('/content/my_images'))
```

# Plot ảnh các thứ nè mà nó bouding box theo phần trăm ko bk plot kiểu j =))

```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
img = cv2.imread('/content/unzip_images/000000000199.jpg', cv2.COLOR_BGR2RGB)
a = np.load('/content/ocr_features/000000000199.npy', allow_pickle=True)
```

```python
for x1, x2, x3, x4 in a.tolist()['boxes']:
  image = cv2.rectangle(img, (int(x1), int(x2)), (int(x3), int(x4)), (255, 0, 0), 1)

cv2_imshow(image)
```
